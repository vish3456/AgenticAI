```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model = "gemini-3.1-flash-lite")
```

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    title: str = Field(
        description="The title of the movie"
    )
    
    year: int = Field(
        description="This year the movie was released"
    )
    
    director: str = Field(
        description="The director of the movie"
    )
    
    rating: float = Field(
        description="The movie's rating out of 10"
    )
    cast: str = Field(
        description="The cast of movie."
    )
```

```python
mws = model.with_structured_output(Movie)
mws
```

```python
mws.invoke("tell me about the inception movie")
```

### Using typedict

```python
from typing_extensions import TypedDict, Annotated

class MovieDict(TypedDict):
    """A movie with details."""

    title: Annotated[
        str,
        ...,
        "The title of the movie"
    ]

    year: Annotated[
        int,
        ...,
        "The year the movie was released"
    ]

    director: Annotated[
        str,
        ...,
        "The director of the movie"
    ]

    rating: Annotated[
        float,
        ...,
        "The movie's rating out of 10"
    ]
```

```python
mwtd = model.with_structured_output(MovieDict)
response = mwtd.invoke("tell me about the avengers.")
response
```
