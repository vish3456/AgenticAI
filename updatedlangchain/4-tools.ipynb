```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model = "gemini-3.1-flash-lite")
response = model.invoke("why does parrot talk??")
response.content[0]['text']
```

```python
from langchain.tools import tool
@tool

def get_weather(location:str)-> str:
    """Get the weather report of a  location"""
    return f"it is sunny in {location}"

model_with_tool = model.bind_tools(get_weather)
```

```python
response = model_with_tool.invoke("what is the weather in the mathura")
for tool_call in response.tool_calls:
    print(f"tool:{tool_call['name']}")
    print(f"Args:{tool_call['args']}")   

```

### Tool Executin Loop

```python

#step1) model generate tool calls
messages = [{'role':'user','content':'what is the weather in the mathura.?'}]
ai_msg = model_with_tool.invoke(messages)
messages.append(ai_msg)

#step2) execute tools and collect result
for tool_call in ai_msg.tool_calls:
    #execute the tool with the generated args.
    tool_result = get_weather.invoke(tool_call)
    messages.append(tool_result)


# Step3) pass result back to  the model
final_response = model_with_tool.invoke(messages)
print(final_response.text)

```
