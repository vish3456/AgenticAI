### Stream

```python
for chunk in model.stream("why  do human at 20s especially men , eager  for love"):
    print(chunk.text, end='|', flush=True)
```

### Batch

```python
responses = model.batch([
    "why do human are so lazy",
    "why do human so close to their religion and god"
    ,
    "what is quantum entanglement"
])

for response in responses:
    print(response)
```

### Stream+Batch

```python
prompts = [
    "why are humans so lazy",
    "why are humans so close to their religion and god",
    "what is quantum entanglement"
]

for prompt in prompts:
    print(f"\nPrompt: {prompt}\n")

    for chunk in model.stream(prompt):
        print(chunk.text, end="", flush=True)

    print("\n" + "-" * 40)
```
